```python
import pandas as pd

# Read US Census data. 
# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# Cluster those data points. 
Z = linkage(df, 'ward')
Z
```

```python
Z.shape
```

```python
# Draw the dendrogram
plt.figure()
dendrogram(Z)
plt.ylabel('Euclidean distance')
plt.show()
```

```python
# Use another distance metrics version. 
Z = linkage(df, 'complete')
Z
```

```python
Z.shape
```

```python
# Draw the dendrogram
plt.figure()
dendrogram(Z)
plt.ylabel('Euclidean distance')
plt.show()
```

```python
# NIPS Conference Papers dataset
# https://archive.ics.uci.edu/ml/datasets/NIPS+Conference+Papers+1987-2015#
# Abstract: This data set contains the distribution of words in the full text of the NIPS conference papers published from 1987 to 2015.
# The document-term matrix was constructed after tokenization, removal of stopwords and truncation of the vocabulary by only keeping words occurring more than 50 times.
```

```python
# Read the data. 
nips = pd.read_csv('../code/data/NIPS_1987-2015.csv')
nips
# Column name: Xyear_paperID
```

```python
nips_transposed = nips.T
nips_transposed
```

```python
# The first row should be the column names.  
nips_transposed.index
```

```python
first_row = nips_transposed.iloc[0]
first_row
```

```python
nips_transposed.columns = first_row
nips_transposed
```

```python
nips_df = nips_transposed.drop('Unnamed: 0')
nips_df
```

```python
# %%timeit -r 1
# 3min 30s Â± 0 ns per loop (mean Â± std. dev. of 1 run, 1 loop each) 
# data size: 134 MB (5811 rows Ã— 11463 columns), CPU: 2.6 GHz Intel Core i7, Memory: 16 GB

# Cluster those papers. 
clusters = linkage(nips_df, 'ward')
```

```python
# %%timeit -r 1
# 1min 23s Â± 0 ns per loop (mean Â± std. dev. of 1 run, 1 loop each) 

# Draw the dendrogram. 
plt.figure()
dendrogram(clusters)
plt.ylabel('Euclidean distance')
plt.show()
```
